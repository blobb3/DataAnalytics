# Correlation analysis

## Libraries and settings

```python
# Libraries
import os
import scipy.stats
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# seaborn graphics settings
sns.set(color_codes=True)

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Show path to the current working directory
print('Current working directory:', os.getcwd())
```

## Import the apartment data

```python
# Define columns
columns = ['web-scraper-order',
           'bfs_number',
           'bfs_name',
           'lat',
           'lon',
           'rooms',
           'area',
           'price',
           'price_per_m2',
           'pop',
           'pop_dens',
           'frg_pct',
           'emp',
           'tax_income']

# Read and select variables
df_orig = pd.read_csv("apartments_data_enriched.csv")[columns]


# Remove missing values
df = df_orig.dropna()
df.head(5)

# Remove duplicates
df = df.drop_duplicates()
df.head(5)
```

## Scatterplot-matrix

```python
sns.set(style="ticks", font_scale=0.8)
g = sns.PairGrid(df[['rooms',
                     'area',
                     'price',
                     'price_per_m2',
                     'pop',
                     'pop_dens',
                     'frg_pct',
                     'emp',
                     'tax_income']], 
                 height=0.9, 
                 aspect=1)
g.map_upper(sns.scatterplot,color='darkgreen', s=8)
g.map_lower(sns.scatterplot, color='darkgreen', s=8)
g.map_diag(plt.hist, color='brown')

#frg_pct = Ausländeranteil = foreigner percentage
#emp = employe = Anteil der Beschäftigten
#pop_dens = population density
#pop = population
```

## Scatterplots and Pearson correlation of selected variables

### Scatterplot price versus area

```python
# Scatterplot (condition is, that it should be linear)
plt.figure(figsize=(7,4))
plt.plot(df['area'], 
         df['price'], 
         'o', 
         markersize=3.5,
         color="green")

# Add title and axes labels
plt.title('Scatterplot price versus area', fontsize=12)
plt.ylabel('price (CHF)', fontsize=12)
plt.xlabel('area (m2)', fontsize=12)

# Set fontsize of tick labels
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)

plt.show()
```

### Pearson correlation price versus area

```python
# Perform Persons correlation
res = scipy.stats.pearsonr(df['area'], df['price'])

# Print result
print(f'Persons correlation coefficient (r): {res[0]:.4f}')
print(f'p-value: {res[1]:.4f}')
```

### Scatterplot price_per_m2 versus area

```python
# Scatterplot - none-linear-relationship (this is not linear yet, but we can transform it with a logarithm transformation)
plt.figure(figsize=(7,4))
plt.plot(df['area'], 
         df['price_per_m2'], 
         'o', 
         markersize=3.5,
         color="green")

# Add title and axes labels
plt.title('Scatterplot price_per_m2 versus area', fontsize=12)
plt.ylabel('price per m2 (CHF)', fontsize=12)
plt.xlabel('area (m2)', fontsize=12)

# Set fontsize of tick labels
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)

plt.show()
```

### Pearson correlation price_per_m2 versus area

```python
# Perform Persons correlation
res = scipy.stats.pearsonr(df['area'], df['price_per_m2'])

# Print result
print(f'Persons correlation coefficient (r): {res[0]:.4f}')
print(f'p-value: {res[1]:.4f}')
```

### Log-transform price_per_m2 and area to make the relationship linear

```python
# Transform values of price_per_m2 to make it linear
#if we use a log, we have at the end a linear relationship (and then we can calculate with it)
#there are also a few outliers (we could take them out with using a subset)
df['log_price_per_m2'] = np.log2(df['price_per_m2'])
df['log_area'] = np.log2(df['area'])

# Scatterplot
plt.figure(figsize=(7,4))
plt.plot(df['log_area'], 
         df['log_price_per_m2'], 
         'o', 
         markersize=3.5,
         color="darkred")

# Add title and axes labels
plt.title('Scatterplot log(price per m2) versus log(area) ', fontsize=12)
plt.ylabel('log(price per m2)', fontsize=12)
plt.xlabel('log(area)', fontsize=12)

# Set fontsize of tick labels
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)

# Add text annotation
plt.text(4, -2.8, s="We have some outliers here ...", fontsize=12, color="blue")

plt.show()
```

### Pearson correlation log(price_per_m2) versus log(area)

```python
# Perform Persons correlation
res = scipy.stats.pearsonr(df['log_area'], df['log_price_per_m2'])

# Print result
print(f'Persons correlation coefficient (r): {res[0]:.4f}')
print(f'p-value: {res[1]:.4f}')
```

## Correlation analysis based on car data

### Import car data

```python
import pandas as pd

# Define columns for import
columns = ['Offer_Id',
          'Type',
          'Init_Regist',
          'Kilometer',
          'Price_Num',
          'PS_Num']

# Read the data and show header info (reads it from the csv-file with almost 4000 tousands datasets)
df_cars = pd.read_csv('autoscout24_data.csv', sep=";")[columns]

# Kilometer to numeric value
df_cars['Kilometer_Num'] = df_cars['Kilometer'].str.replace(' km', '').str.replace('\'', '').astype('int64')

# Remove original column 'Kilometer' 
df_cars = df_cars.drop(columns=['Kilometer'])

# Rename columns
df_cars = df_cars.rename(columns={'Price_Num':'Price', 'PS_Num':'PS', 'Kilometer_Num':'Kilometer'})
print('Number of rows and columns:', df_cars.shape)

# print header info
df_cars
```

```python
g = sns.PairGrid(df_cars, diag_sharey=False, corner=True)
g.map_lower(sns.scatterplot,color='orange', s=8)
g.map_diag(sns.kdeplot, color='red')
```

```python
g = sns.PairGrid(df_cars, diag_sharey=False)
g.map_diag(sns.histplot, color='black')
g.map_offdiag(sns.scatterplot, color='red')
g.add_legend()
```

```python
g = sns.PairGrid(df_cars, diag_sharey=False)
g.map_upper(sns.scatterplot, color='darkgreen')
g.map_lower(sns.kdeplot, color='lightgreen')
g.map_diag(sns.kdeplot, color='green')
```

```python
# Perform Persons correlation
res = scipy.stats.pearsonr(df_cars['Price'], df_cars['Kilometer'])

#print result (Unterschied von 0 = hoch signifikant)
# f'p-value: {res[1]:.4f} ==> das ist ein ein String mit bestimmten Formeln formatieren
print(f'Persons correlation coefficient (r): {res[0]:.4f}')
print(f'p-value: {res[1]:.4f}')

f'p-value: {res[1]}' #Ist r < 0 so besteht ein negativer Zusammenhang. 
```

```python
# Perform Persons correlation
res = scipy.stats.pearsonr(df_cars['Price'], df_cars['PS'])

#print result (Unterschied von 0 = hoch signifikant)
# f'p-value: {res[1]:.4f} ==> das ist ein ein String mit bestimmten Formeln formatieren
print(f'Persons correlation coefficient (r): {res[0]:.4f}')
print(f'p-value: {res[1]:.4f}')

f'p-value: {res[1]}' #Ist r>0 so besteht ein positiver Zusammenhang.
```

```python
# Perform Persons correlation
res = scipy.stats.pearsonr(df_cars['Kilometer'], df_cars['PS'])

#print result (Unterschied von 0 = hoch signifikant)
# f'p-value: {res[1]:.4f} ==> das ist ein ein String mit bestimmten Formeln formatieren
print(f'Persons correlation coefficient (r): {res[0]:.4f}')
print(f'p-value: {res[1]:.4f}')

f'p-value: {res[1]}' #hier ist sie sehr klein, deshalb ist ein Zusammenhang sicher
```
