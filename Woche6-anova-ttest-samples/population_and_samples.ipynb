# Population and samples in statistics

## Libraries and settings

```python
# Libraries
import numpy as np
import random
import statistics
import matplotlib.pyplot as plt

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

## Adult residents in the city of ZÃ¼rich with their body heights (population)

```python
# Setting a seed to restore the random numbers later
np.random.seed(10)

# Generate (normally distributed) data with mean = 180 cm and standard deviation = 10 cm
pop = np.random.normal(loc=180, scale=10, size=348000)
print(f'{len(pop):.0f}')
print(f'{np.mean(pop):.4f}')
print(f'{np.std(pop):.4f}')

# Plot histogram based on the data
import matplotlib.pyplot as plt
plt.hist(pop, bins = 50, color='green')
plt.title('Body heights of adults in the city of Zuerich')
plt.xlabel('Body height (cm)')
plt.ylabel('Number of residents')

plt.show()
```

## Single random sample from the population

```python
# Sample size
sample_size = 4000

# Setting a seed to restore the random numbers later
random.seed(10)

# Random sample
random_sample = random.sample(list(pop), sample_size)

# Statistics of the sample
# The mean is the sum of all the entries divided by the number of entries.
print(f'Mean: {statistics.mean(random_sample):.4f}')
# Standard deviation is a measure of the amount of variation or dispersion of a set of values. 
#We first need to calculate the mean of the values, then calculate the variance, and finally the standard deviation.
print(f'Standard deviation: {statistics.stdev(random_sample):.4f}')

# Plot histogram based on the sample
import matplotlib.pyplot as plt
plt.hist(random_sample, bins = 50, color='deepskyblue')
plt.title(f'Body height of adults in the city of Zuerich (sample with n={sample_size})', fontsize=10)
plt.xlabel('Body height (cm)')
plt.ylabel('Number of residents')

plt.show()
```

## Example 

```python
# Sample size
sample_size = 200

# Setting a seed to restore the random numbers later
random.seed(10)

# Random sample
random_sample = random.sample(list(pop), sample_size)

# Statistics of the sample
print(f'Mean: {statistics.mean(random_sample):.4f}')
print(f'Standard deviation: {statistics.stdev(random_sample):.4f}')

# Plot histogram based on the sample
import matplotlib.pyplot as plt
plt.hist(random_sample, bins = 50, color='red')
plt.title(f'Diameters of trees in a forest (sample with n={sample_size})', fontsize=10)
plt.xlabel('Tree Diameter (mm)')
plt.ylabel('Number of trees')

plt.show()
```

## Repeated random sampling 'with replacement' (bootstrapping)

```python
# Define number of samples and sample size
num_iterations = 10000
sample_size = 200

# Create empty lists to save results of iterations
sample_mean = []
sample_std  = []

# Loop for iterations
for i in range(num_iterations):
    samp = random.sample(pop.tolist(), sample_size)
    avg = np.mean(samp)
    std = np.std(samp)
    sample_mean.append(avg)
    sample_std.append(std)

# Print mean and sd of bootsstapped distribution
print(f'{np.mean(sample_mean):.4f}')
print(f'{np.mean(sample_std):.4f}')

# Plot histogram based on the bootstrapped data
plt.hist(sample_mean, bins = 50, color='orange')
plt.title('Body height of adults in the city of Zuerich (bootstrap sample)', fontsize=10)
plt.xlabel('Body height (cm)')
plt.ylabel('Number of residents')
plt.show()
```

#### Aufgabe d
As sample sizes increase, the sampling distributions approach a normal distribution. With "infinite" numbers of successive random samples, the mean of the sampling distribution is equal to the population mean (Âµ).
