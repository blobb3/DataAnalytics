# Getting all available Swiss electronic stores using overpass Web API

## Overpass turbo

#### Why electronics shops? 

Since we scraped our data from Amazon (we limited ourselves to laptops), we came up with the idea to analyse the competition in Switzerland, as online providers increasingly stealing customers from the local shops.

#### Approach with overpass turbo

We approached the Overpass API by interactively trying out our own queries on the http://overpass-turbo.eu/ website. 
The results are clearly displayed on a map. The electronics shops in Switzerland are obtained by navigating to Switzerland on the map and entering the data "elektronikfachgeschfte under Wizard. 
After that, we just copy the code from the terminal and paste it into Jupyter Notebook - after we have written in the URL for the Overpass API.



we use the thre basic components, which are nodes, ways and relations. Nodes are points on the maps (in latitude and longitude). A way is a ordered list of nodes, which correspond (in our example) to the outline of house/store. And the data element relation, which is also an ordered list containing either nodes, ways or even other relations, is used for large structures - in our example to stores, which contain multiple polygons to describe the stores.
At first we wanted to query nodes for the electronic stores etc. - we also applied a filter by tag on our query which looks for all the nodes, where the key-value pair is ("shop"="electronics"). After we spefied the output of the queried data, we make the Web API request.


#### Overpass turbo query to get all available electronc stores in Switzerland

```python
# import statements
from pandas import json_normalize
import pandas as pd
import requests
import json

# Overpass API URL
url = "http://overpass-api.de/api/interpreter"

# Overpass turbo query - we work with the basic components, which are nodes, ways and relations
query = f"""[out:json][timeout:25];
        area["ISO3166-1"="CH"][admin_level=2];
        (node["shop"="electronics"](area); 
          way["shop"="electronics"](area);
          relation["shop"="electronics"](area);
        );
        out body;
        >;
        out skel qt;"""


# Web API request
r = requests.get(url, params={'data': query})
data = r.json()['elements']


# Save data to file
with open('amazon_laptops.json', 'w') as json_file:
    json.dump(data, json_file)

# Store data in data frame
df = json_normalize(data)
df.head(10)
```

## Plot electornics strores on map

Since we wanted to know in which area of Switzerland most electronics shops are located, we first tried to get a rough overview by evaluating the coordinates.
Since we used the "out center;" statement in this query, we get for each way and relation a center coordinate stored under the center key. In the case of node elements, the coordinates are simply under the lat, lon keys. Then we converted our coordinates into a numpy array. With that, our assumption, that most shops are located in densely populated regions, was confirmed. 

Of course, the coordinates were still too imprecise, so we wanted to mark the best-known electronics shops on a map.


#### First rough overview by imported corrdinates

```python
import overpy
import numpy as np
import matplotlib.pyplot as plt

# Access the overpass API with Python by using the overpy package as wrapper.
api = overpy.Overpass()
r = api.query("""
        area["ISO3166-1"="CH"][admin_level=2];
        (node["shop"="electronics"](area);
        way["shop"="electronics"](area);
        relation["shop"="electronics"](area););
        out center; 
        """)

# Collect coordinates into list
coords  = []
coords += [(float(node.lon), float(node.lat)) 
           for node in r.nodes]
coords += [(float(way.center_lon), float(way.center_lat)) 
           for way in r.ways]
coords += [(float(rel.center_lon), float(rel.center_lat)) 
           for rel in r.relations]

    
# Convert coordinates into numpy array
X = np.array(coords)

plt.plot(X[:, 0], X[:, 1], 'o')
plt.title('Electronic stores in Switzerland')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.axis('equal')
plt.show()
```

#### Plot the electronic stores on a map

```python
import folium
import pandas as pd

# Subset of electronic stores by brand
locations = df[["lat", "lon", "tags.brand", "tags.shop"]].loc[(df["tags.brand"] == 'Interdiscount') | 
                                                              (df["tags.brand"] == 'Fust') |
                                                              (df["tags.brand"] == 'melectronics') |
                                                              (df["tags.brand"] == 'Media Markt ') |
                                                              (df["tags.brand"] == 'Digitec')    
                                                             ] 
print(locations.head(20))

# Create map
map = folium.Map(location=[locations.lat.mean(), 
                           locations.lon.mean()], 
                 zoom_start=8, 
                 control_scale=True)

# Add maker symbols
for index, location_info in locations.iterrows():
    folium.Marker([location_info["lat"], 
                   location_info["lon"]], 
                  popup=location_info["tags.brand"]).add_to(map)

# Plot map
map
```
