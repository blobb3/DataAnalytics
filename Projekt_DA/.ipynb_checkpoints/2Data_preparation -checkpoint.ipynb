#### Extract dimensions

```python
dimension = []
for i in df['dimensions_raw']:
    d1 = re.findall('(.*)cm', str(i))
    try:
        d2 = d1[0].strip().replace(',', '.')
    except:
        d2 = None
    dimension.append(d2)

# Save as new variable in the pandas data frame
df['dimension'] = pd.Series(dimension, dtype="string")
    
# Print first 5 values
print(df['dimensions_raw'].head(5), '\n')
print(df['dimension'].head(5))
```

# Simple linear regression (only one explanatory variable in the model)

```python
#Create train and test samples (train = 80%, test = 20% of the data)
X_train, X_test, y_train, y_test = train_test_split(df['size'], 
                                                    df['price'], 
                                                    test_size=0.20, # 80:20 
                                                    random_state=42)
print('X_train:')
print(X_train.head(), '\n')
print('y_train:')
print(y_train.head())
```

```python
# Fit the regression model mit Linearer Einfach-Regression
slope, intercept, r, p, std_err = stats.linregress(X_train, y_train)

# Print results of the regression model (slope = Steigung der Regressionsgerade)
print('Linear regression result:')
print(f'Intercept with y-axis (alpha):            {intercept:.2f}') # intersection with y axis, if R=0
print(f'Slope of regression line (beta):          {slope:.3f}') # m = 28.583 => with each unit my x increases, the price (y) increases by 28.6 (= marginal effect)
print(f'p-value:                                  {p:.4f}')
print(f'R-squared (coefficient of determination): {r**2:.4f}') # A value of 0.5246 means, that the explanatory variables explain 52.46% of the variaton of our target variable (laptop prices) - not bad, but could be improved.
```

```python
# Function to calculate model predictions
def myfunc(x):
    return slope * x + intercept

# Apply myfunc() to x, i.e. make predictions 
mymodel = pd.Series(map(myfunc, X_train))

# Scatterplot with regression line
plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train, s=60, color='lightgreen')
plt.plot(X_train, mymodel, color='red', linestyle='dashed')
plt.title('Simple Linear Regression')
plt.xlabel('size')
plt.ylabel('price(Euro)')

plt.show()
```

```python
# Calculate residuals (here: predicted prices minus observed prices)
residuals = mymodel - y_train

# Check the first residual value in our data set
print(f'1st Predicted price in dataset: {mymodel[0]:.2f}')
print(f'1st Observed price in dataset: {y_train.iloc[0]:.2f}')
print(f'1st Residual price in dataset: {residuals[0]:.2f}')
```

### Classification using Decision trees

```python
# Step 1: Import the libraries
import sklearn
from sklearn import datasets
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier # for classification problems
from sklearn.tree import DecisionTreeRegressor # for regression problems
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt
import seaborn as sns
    
# Step 2 - Setup the Data for classifier => we have created objects X and y to store the data and the target value respectively.
# train_test_split => will give us four different outcomes
X_train, X_test, y_train, y_test = train_test_split(df[['size']], 
                                                    df[['price']], 
                                                    test_size=0.20, # 80:20 
                                                    random_state=42)

# Step 3 - Model and its Score 
# => we are using DecisionTreeClassifier as a Machine Learning model to fit the data
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
print(model)

# Step 4 - we predicted the output by passing X-test 
# => we now print the classification report and confusion matrix for the classifier
# make predictions
expected_y  = y_test #y_test-predictions
predicted_y = model.predict(X_test)  
print('\n Classification Report: \n', metrics.classification_report(expected_y, predicted_y)) #gives a complete report of our example in context of classification , from the confusion matrix we get the values: precison, recall, f1-score, support
print('\n Confusion Matrix: \n', metrics.confusion_matrix(expected_y, predicted_y))



# summarize the fit of the model
print('\n R2_score: ',metrics.r2_score(expected_y, predicted_y))
print('\n Mean squared log error for the regressor/Log Score: ',metrics.mean_squared_log_error(expected_y, predicted_y))
# mean_squared_log_error => it will make a mean_square and take the log of those values and sum it up for us
```
